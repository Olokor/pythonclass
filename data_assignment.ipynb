```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_os, y_os, test_size=0.2, random_state=42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# ML models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score, accuracy_score, classification_report

models = [
    LogisticRegression(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    GradientBoostingClassifier(),
    SVC()
]


```

```python
for model in models:
    print(f"{model.__class__.__name__} performance................")

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    c_matrix = confusion_matrix(y_test, y_pred)
    print(f"for {model.__class__.__name__} model, accuracy_score is:{accuracy_score(y_test, y_pred)}")
    print(classification_report(y_test, y_pred))

    fig = plt.figure(figsize=(15, 6))

    ax1 = fig.add_subplot(1, 2, 1)
    ax1 = sns.heatmap(pd.DataFrame(c_matrix), annot=True, cmap="Blues", fmt='d')
    bottom, top = ax1.get_ylim()
    ax1.set_ylim(bottom + 0.5, top - 0.5)
    plt.xlabel("Predicted")
    plt.ylabel("Expected")

    ax2 = fig.add_subplot(1, 2, 2)
    if hasattr(model, 'predict_proba'):
        y_pred = model.predict(X_test)
        y_pred_proba = model.predict_proba(X_test)[:, 1] #RETURNING A 2D
    elif hasattr(model, 'predict'):
        y_pred = model.predict(X_test)
        y_pred_proba = y_pred
    else:
        continue
    fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
    auc = roc_auc_score(y_test, y_pred_proba)
    ax2 = plt.plot(fpr, tpr, label=f"AUC={auc:.2f}")
    plt.plot([0, 1], [0, 1], 'k--')
    plt.legend(loc='lower right')
    plt.show()
```

# hypothesis testing

```python
from scipy import stats
# Separate data into two groups based on Attrition
data = new_df.copy()
attrition_yes = data[data['Attrition'] == 1]
attrition_no = data[data['Attrition'] == 0]

# Function to perform t-test for continuous variables
def t_test(group1, group2, column):
    return stats.ttest_ind(group1[column], group2[column], nan_policy='omit')

# Function to perform chi-square test for categorical variables
def chi_square_test(group1, group2, column):
    contingency_table = pd.crosstab(data[column], data['Attrition'])
    return stats.chi2_contingency(contingency_table)

# Columns to test
continuous_columns = ['Age', 'DailyRate', 'DistanceFromHome', 'HourlyRate', 
                      'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked', 
                      'PercentSalaryHike', 'TotalWorkingYears', 
                      'TrainingTimesLastYear', 'YearsAtCompany', 
                      'YearsInCurrentRole', 'YearsSinceLastPromotion', 
                      'YearsWithCurrManager']

categorical_columns = ['BusinessTravel', 'Department', 'Education', 'EducationField', 
                       'EnvironmentSatisfaction', 'Gender', 'JobInvolvement', 
                       'JobLevel', 'JobRole', 'JobSatisfaction', 'MaritalStatus', 
                       'Over18', 'OverTime', 'PerformanceRating', 
                       'RelationshipSatisfaction', 'StockOptionLevel', 
                       'WorkLifeBalance']

results = {}

for column in continuous_columns:
    results[column] = t_test(attrition_yes, attrition_no, column)

for column in categorical_columns:
    results[column] = chi_square_test(attrition_yes, attrition_no, column)

# Display the results
for column, result in results.items():
    if column in continuous_columns:
        print(f'{column}: t-statistic = {result.statistic}, p-value = {result.pvalue}')
    else:
        print(f'{column}: chi2 = {result[0]}, p-value = {result[1]}')
```

```python
# confirm the test
for column, result in results.items():
    if result.pvalue < 0.05: #5% tolerance value
        print(f"we can reject the null hypothesis for {column} column")
    else:
        print(f"we can accept the null hypothesis for {column} column")
```

```python
import statsmodels.stats.proportion as smp

# Create groups based on travel frequency
group_A = data[data['BusinessTravel'] == 1]
group_B = data[data['BusinessTravel'] == 2]

# Calculate attrition rates for each group
attrition_rate_A = group_A[group_A['Attrition'] == 1]['Attrition'].count() / len(group_A)
attrition_rate_B = group_B[group_B['Attrition'] == 1]['Attrition'].count() / len(group_B)

# Perform A/B test
count_A = group_A[group_A['Attrition'] == 1]['Attrition'].count()
nobs_A = len(group_A)
count_B = group_B[group_B['Attrition'] == 1]['Attrition'].count()
nobs_B = len(group_B)

z_score, p_value = smp.proportions_ztest([count_A, count_B], [nobs_A, nobs_B], alternative='two-sided')

print(f'Attrition rate for Travel_Frequently group: {attrition_rate_A:.2f}')
print(f'Attrition rate for Travel_Rarely group: {attrition_rate_B:.2f}')
print(f'z-score: {z_score:.2f}')
print(f'p-value: {p_value:.4f}')

if p_value < 0.05: #tolerance of 5%
    print('The attrition rates are significantly different (p < 0.05).')
else:
    print('The attrition rates are not significantly different (p >= 0.05).')
```
